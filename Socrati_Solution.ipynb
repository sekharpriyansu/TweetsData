I am thinking to predict the number of rentals during each hour of the day, so it makes sense to split the datetime column into 'day of the week' and 'month of the year' as we think these might be useful features for the model and might provide insights.

## Transformed Linear Regression

From the above analysis, we realise that modeling Y as a linear function of the regression coefficients may not be sensible and a more realistic model might be to use the natural log of the response variable.

In order to get the log of 'vals' column we have to make sure that the vals column doesn't have any 0 values.

Lets find the minimum of vals column.

```python
finalDF.describe()
```

From the above description, minimum value of vals column is 1. Hence, it is ok to just directly convert to log(y) as min(y) is 1 and not 0 in the 'vols' column.

```python
finalDF_mod = finalDF
#Build the first model using total vals as a label so convert the vals column to log vols
finalDF_mod['log_vals'] = finalDF['vals'].map(lambda x: np.log(x))
#Remove the vols
finalDF_mod = finalDF.drop('vals', 1)
```

Let's check the new DataFrame out.

```python
finalDF_mod.head()
```

Lets create a function to calculate Root Mean Squared Logarithmic Error (RMSLE), which would provide better understandig on error. 

```python
def rmsele(actual, pred):
    squared_errors = (np.log(pred + 1) - np.log(actual + 1)) ** 2
    mean_squared = np.sum(squared_errors) / len(squared_errors)
    return np.sqrt(mean_squared)
```

Now,we will create a train and test split with logarithmic lables and check our findings.

```python
x_data = finalDF_mod.drop('log_vals', 1)
print x_data.shape
y_data = finalDF_mod['log_vals']

xtrain, xtest, ytrain, ytest = train_test_split(x_data, y_data)
clf = LinearRegression()
clf.fit(xtrain, ytrain)

#Root Mean Squared Logarithmic Error (RMSLE)
print "Root Mean Squared Logarithmic Error: ", rmsele(np.exp(clf.predict(xtest)), np.exp(ytest))
print "Training accuracy: %0.2f%%" % (100*clf.score(xtrain, ytrain))
print "Test accuracy: %0.2f%%" % (100*clf.score(xtest, ytest)) + "\n"
```

From the above results, it is clear that this model out performed our previous model by miles. This model has good traing and test accuracy which is above 90%. However, RMSLE value is still little high and we may have very distant values or outliers.

Let's check out our Poisson Regression Model Prediction

```python
plt.scatter(np.exp(clf.predict(xtest)), np.exp(ytest))
plt.title("Poisson Regression Model Prediction")
plt.ylabel("Actual vals")
plt.xlabel("Predicted vals")
plt.xlim(0, 3200)
plt.ylim(0, 3500)
```

From the above plot, it is clear that the data is showing linearity except few outliers. We can confirm that the model appears to capture the general trends in the data

##  Improved Model

### Random Forest

We now implement random forest models to reduce the root mean squared logarithmic error. For this section we will use the non 'hot-encoded' dataframe, as the random forest will not be affected by using this data.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
from sklearn.metrics import make_scorer

finalDF_mod = finalDF_mod.drop('log_vals', 1)
y_data = data['vals']
x_data = finalDF_mod
features = x_data.columns

xtrain, xtest, ytrain, ytest = train_test_split(x_data, y_data)
```

#### Now we run a grid search algorithm in order to optimise the model parameters.

```python
from sklearn.grid_search import GridSearchCV

# This helper function will make a callable that we can use in cross_val_score
rmsele_scorer = make_scorer(rmsele, greater_is_better=False)

#Parameter Optimisation with grid search
tuned_parameters = [{'max_features': ['sqrt', 'log2', 'auto'], 'max_depth': [5, 8, 12], 'min_samples_leaf': [2, 5, 10]}]
rf =  GridSearchCV(RandomForestRegressor(n_jobs=-1, n_estimators=1000), tuned_parameters, cv=3, verbose=2, scoring=rmsele_scorer).fit(xtrain, ytrain)
```

```python
print 'Best Parameters:'
print rf.best_estimator_
```

RandomForestRegressor is trained using the best parameters.

```python
rf_op = RandomForestRegressor(bootstrap=True, compute_importances=None,
           criterion='mse', max_depth=12, max_features='auto',
           max_leaf_nodes=None, min_density=None, min_samples_leaf=2,
           min_samples_split=2, n_estimators=1000, n_jobs=-1,
           oob_score=False, random_state=None, verbose=0)
rf_op=rf_op.fit(x_data, y_data)
```

```python
numfeat = len(features)
indices = np.argsort(rf_op.feature_importances_)[::-1][:numfeat]
plt.bar(xrange(numfeat), rf_op.feature_importances_[indices], align='center', alpha=.5)
plt.xticks(xrange(numfeat), features[indices], rotation='vertical', fontsize=12)
plt.xlim([-1, numfeat])
plt.ylabel('Feature percentages', fontsize=12)
plt.title('Feature importances computed by Random Forest', fontsize=16)
print zip(features, rf_op.feature_importances_)
```

```python
#Root Mean Squared Logarithmic Error (RMSLE)
print "Root Mean Squared Logarithmic Error Train: ", rmsele(rf_op.predict(xtrain), ytrain)
print "Root Mean Squared Logarithmic Error Test: ", rmsele(rf_op.predict(xtest), ytest)
print "Training accuracy: %0.2f%%" % (100*rf_op.score(xtrain, ytrain))
print "Test accuracy: %0.2f%%" % (100*rf_op.score(xtest, ytest)) + "\n"
```

From the above scores, it is clear that this model is performing poor and not predicting the tests as well.
